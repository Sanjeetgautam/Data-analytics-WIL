```python

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder



# Define a dictionary to map categorical values to numerical values
mapping = {
    'Contract': {'Month-to-month': 0, 'One year': 1, 'Two year': 2},
    'InternetService': {'DSL': 0, 'Fiber optic': 1, 'No': 2},
    'Churn': {'No': 0, 'Yes': 1},
    'MultipleLines': {'No phone service': 0, 'No': 1, 'Yes': 2},
    'PhoneService': {'No': 0, 'Yes': 1},
    'Dependents': {'No': 0, 'Yes': 1},
    'gender': {'Female': 0, 'Male': 1}
}

# Use the 'replace' method to map the categorical values
for column, value_map in mapping.items():
  data[column] = data[column].replace(value_map)

data.head()
```

```python
# Separate features (X) and target (y)
X = data.drop('Churn', axis=1)
y = data['Churn']

```

```python
scaler = StandardScaler()
features_scaled = scaler.fit_transform(X)
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features_scaled, y, test_size=0.2, random_state=42)

```

```python
# prompt: make a csv files from train_set.csv
# test_set.csv

import pandas as pd

# Assuming X_train and y_train are your training data and labels
train_data = pd.DataFrame(X_train)
train_data['Churn'] = y_train

# Assuming X_test and y_test are your testing data and labels
test_data = pd.DataFrame(X_test)
test_data['Churn'] = y_test

# Save the dataframes to CSV files
train_data.to_csv('train_set.csv', index=False)
test_data.to_csv('test_set.csv', index=False)
```
